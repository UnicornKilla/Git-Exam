Задача 2.
Напишите программу, которая принимает на вход число x.
Если:

x > 0 → вывести "Положительное",

x < 0 → "Отрицательное",

иначе "Ноль".

Если число чётное — добавьте к выводу " (чётное)",
иначе " (нечётное)".

````python
number = int( input("Введите число: ") )


if number < 0:
    print("Отрицательное число")
    
    if number % 2 == 0:
        print("Чётный")
    else:
        print("Нечётный")
elif number > 0:
    print("Положительное число")

    if number % 2 == 0:
        print("Чётный")
    else:
        print("Нечётный")
else:
    print("Ноль")
    print("Чётный")

````

Задача 3.
Есть список покупок:

purchases = [350, 1200, 700, 400, 1000, 250]

- Найдите сумму всех покупок.
- Найдите среднюю покупку.
- Сформируйте новый список из покупок больше 500.
- Выведите количество таких покупок.

````python
purchases = [350, 1200, 700, 400, 1000, 250]

summa_spiska = sum(purchases)
print("Сумма: ", summa_spiska)

srednee = summa_spiska / len(purchases)
print("Среднее число: ", srednee)

bolee_500 = []

for i in purchases:
    if i > 500:
        bolee_500.append(i)

print("Список с продажами более 500: ", bolee_500)

print(len(bolee_500))
````

Задача 4. Словари и множества (20 баллов)

Есть словарь:

students = {
    "Анна": 44,
    "Иван": 92,
    "Мария": 53,
    "Павел": 85,
    "Ольга": 97
}


- Выведите имена студентов, у которых оценка выше 80.
- Выведите имена студентов, у которых оценка меньше 60.
- Найдите среднюю оценку по классу.


````python
students = {
    "Анна": 44,
    "Иван": 92,
    "Мария": 53,
    "Павел": 85,
    "Ольга": 97
}

bolee_80 = []
menee_60 = []

for name, mark in students.items():

    if mark > 80:
        bolee_80.append(name)
        print(f"У {name} оценка больше 80")

    elif mark < 60:
        menee_60.append(name)
        print(f"У {name} оценка меньше 60")

srednee_studentov = sum(students.values()) / len(students.values())

print("Среднее студентов", srednee_studentov)
print("Более 80", bolee_80)
print("Менее 60", menee_60)
````

````python
data = (4, 8, 15, 16, 23, 42)

def analyze_numbers(data):
    chetnie_count = 0

    for i in data:
        if i % 2 == 0:
            chetnie_count += 1
    

    return {
        "min": min(data), 
        "max": max(data), 
        "average": sum(data) / len(data),
        "even_count": chetnie_count
    }

print(analyze_numbers(data))
````

Задача 6.
“Анализ продаж магазина”

Создайте программу с функциями, которая анализирует словарь:
```py
sales = {
  "Понедельник": 42000,
  "Вторник": 38000,
  "Среда": 50000,
  "Четверг": 47000,
  "Пятница": 53000,
  "Суббота": 61000,
  "Воскресенье": 29000
}
```

Функция sales_report(data) должна вернуть словарь:
```py
{
  "total": ... ,              # общая выручка за неделю
  "average": ... ,            # средняя выручка
  "best_day": ... ,           # день с максимальной выручкой
  "worst_day": ... ,          # день с минимальной выручкой
  "weekend_share": ...        # доля выходных в % от общей выручки
}
```

Пример результата:
```py
{
  "total": 320000,
  "average": 45714.3,
  "best_day": "Суббота",
  "worst_day": "Воскресенье",
  "weekend_share": 28.1
}
```



````python
sales = {
  "Понедельник": 42000,
  "Вторник": 38000,
  "Среда": 50000,
  "Четверг": 47000,
  "Пятница": 53000,
  "Суббота": 61000,
  "Воскресенье": 29000
}

def sales_report(data):
    
    best_day = ''
    best = 0

    worst = max(data.values())
    worst_day = ''

    for name, sale in data.items():
        print(sale)
        if best < sale:
            best = sale
            best_day = name
        
        if worst > sale:
            worst = sale
            worst_day = name

        weekend_share = ( (data['Суббота'] + data['Воскресенье']) / sum(data.values()) ) * 100


    result = {
      "total": sum(data.values()),
      "average": sum(data.values()) / len(data.values()),
      "best_day": best_day,
      "worst_day": worst_day,
      "weekend_share": weekend_share
    }

    return result

print(sales_report(sales))
````

# Слияние для Километрожа Входа

````python
stops_df
````

````python
stops_df_entry = stops_df.rename(columns={
    "Остановки": "Остановка входа",
    "Километры": "Километры_Вход"
})

stops_df_entry
````

````python
df_105 = pd.merge(
    df_105,
    stops_df_entry[ ['Остановка входа', 'Километры_Вход'] ],
    how = 'left',
    on = 'Остановка входа'
)
````

# Слияние для Километрожа Выхода

````python
stops_df_exit = stops_df.rename(columns={
    "Остановки": "Остановка выхода",
    "Километры": "Километры_Выход"
})

df_105 = pd.merge(
    df_105,
    stops_df_exit[ ['Остановка выхода', 'Километры_Выход'] ],
    how = 'left',
    on = 'Остановка выхода'
)

df_105
````
